```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
all_df = pd.read_csv("mnist_all_accuracy.csv", index_col=0)
```

```python
X=all_df.iloc[:,:]#将0到61列数据赋值给X
X=X.values#.values方法将dataframe转为numpy.ndarray，也可以用np.array(X)将其转为numpy.ndarray
X.tolist()
```

```python
for temp_list in X:
    for i in range(len(temp_list)):
        temp_list[i] = eval(temp_list[i])
```

```python
X_list = list(X)
```

```python
for i in range(len(X_list)):
    temp_list = list(X_list[i])
    temp_df = pd.DataFrame(temp_list)
    temp_df.to_csv("random/mnist/0."+str(i)+".csv")
```
